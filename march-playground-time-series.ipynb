```python
import numpy as np # linear algebra
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
from pylab import rcParams
rcParams["figure.figsize"] = 20,5
```

```python
data=pd.read_csv("../input/tabular-playground-series-mar-2022/train.csv",parse_dates=['time'],index_col='row_id')
test=pd.read_csv("../input/tabular-playground-series-mar-2022/test.csv",parse_dates=['time'],index_col='row_id')
submission=pd.read_csv("../input/tabular-playground-series-mar-2022/sample_submission.csv")
plt.plot(data[100:150]['congestion'])
plt.show()
```

```python
rolling_series=data[1:50]['congestion'].rolling(window=43)
rolling_mean=rolling_series.mean()
rolling_mean.plot(color='red')
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
adfuller_result = adfuller(data[1:100].congestion.values, autolag='AIC')
print(f"Stats---{adfuller_result[0]}")
print(f"P-values---{adfuller_result[1]}")
```

```python
for df in [data, test]:
    df['weekday'] = df.time.dt.weekday
    df['hour'] = df.time.dt.hour
    df['minute'] = df.time.dt.minute
    
```

```python
medians = data.groupby(['x', 'y', 'direction', 'weekday', 'hour', 'minute']).congestion.mean().astype(int)
medians
```

```python
test.head()
```

```python
final=test.merge(medians, 
                 left_on=['x', 'y', 'direction', 'weekday', 'hour', 'minute'],
                 right_index=True)[['congestion']]
final.reset_index(inplace=True)
final
```

```python
final.to_csv('submission.csv',index=False)
```
